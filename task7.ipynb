```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import classification_report
```

```python
file_path = "C:/Users/JAAVANIKA L/fall semester 22-23/Downloads/breast-cancer.csv"
df = pd.read_csv(file_path)
print("First 5 rows of your dataset:")
df.head()
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
svm_linear = SVC(kernel='linear', C=1)
svm_linear.fit(X_train, y_train)
svm_rbf = SVC(kernel='rbf', C=1, gamma='scale')
svm_rbf.fit(X_train, y_train)
```

```python
def plot_decision_boundary(model, X, y, title):
    h = 0.02
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(6, 4))
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, edgecolors='k')
    plt.title(title)
    plt.xlabel(X.columns[0])
    plt.ylabel(X.columns[1])
    plt.show()

plot_decision_boundary(svm_linear, X, y, "SVM with Linear Kernel")
plot_decision_boundary(svm_rbf, X, y, "SVM with RBF Kernel")"
```

```python
param_grid = {
    'C': [0.1, 1, 10],
    'gamma': [1, 0.1, 0.01],
    'kernel': ['rbf']
}
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=1, cv=5)
grid.fit(X_train, y_train)
print("Best Parameters:", grid.best_params_)
```

```python
scores = cross_val_score(grid.best_estimator_, X, y, cv=5)
print("Cross-validation scores:", scores)
print("Average accuracy:", scores.mean())

y_pred = grid.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
